```python
import numpy as np
import pandas as pd
```

```python
#importing data frame
df_health=pd.read_csv('dailyActivity_merged.csv')
df_health
```

```python
df_health.nunique()
```

```python
#findig null values
df_health.isnull().sum()
```

```python
#checking for data type error
df_health.info()
```

```python
#converting data type of ActivityDate from object to datetime
df_health["ActivityDate"]=pd.to_datetime(df_health["ActivityDate"])
```

```python
df_health.info()
```

```python
#checking if there is any null value
df_health.duplicated().sum()
```

```python
df_health.describe()
```

```python
#creating a new column of TotalMinutes by adding VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes 
df_health['TotalMinutes']=df_health['VeryActiveMinutes']+df_health['FairlyActiveMinutes']+df_health['LightlyActiveMinutes']+df_health['SedentaryMinutes']
df_health
```

```python
#creating a new column %SedentaryMinutes to calculate inactive populatin by SedentaryMinutes/totalminutes*100 formual 
df_health['%SedentaryMinutes']=df_health['SedentaryMinutes']/df_health['TotalMinutes']*100
df_health
```
